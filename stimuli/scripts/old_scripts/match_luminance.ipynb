```python
from skimage import exposure
from skimage import data
from skimage.exposure import match_histograms
import matplotlib.pyplot as plt
import argparse
import cv2
import numpy as np
from PIL import Image
```

```python
src = cv2.imread('/Users/chrisiyer/Downloads/id001_00C_C.jpg')
ref = cv2.imread('/Users/chrisiyer/Downloads/id002_00C_C.jpg')

matched = exposure.match_histograms(src, ref, multichannel=True)
```

```python
plt.imshow(src)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(src.astype('uint8'))
ax1.set_title('Source')
ax2.imshow(ref.astype('uint8'))
ax2.set_title('Reference')
ax3.imshow(matched.astype('uint8'))
ax3.set_title('Matched')

plt.tight_layout()
plt.show()

# Image.fromarray(reference).show()
```
